```python
# DateTime
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d %H:%M:%S')
file = data_list[1]
data = pd.read_csv(mypath + file, parse_dates=['dattim'], date_parser=dateparse, engine='python', index_col='dattim')
```

# 3) Feature Engineering

```python
data1 = pd.DataFrame(data['avepow'])
```

```python
for i in range(1, 14):
    value = data['avepow'].shift(periods=i)
    data1['avepow_t-' + str(i)] = value
    #data1['avepow_t-' + str(i) + '_quad'] = value ** 2
    #data1['avepow_t-' + str(i) + '_cubic'] = value ** 3
    

data1.dropna(axis=0, inplace=True, )
```
